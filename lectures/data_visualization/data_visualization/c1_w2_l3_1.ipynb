## Text properties

#### Formulas in TEX format

```python
plt.title('Quadratic function')
plt.text(0, 75, '$y=x^2$', fontsize = 16)
plt.plot([x**2 for x in range(10)])
```

#### Plot legend

```python
plt.title('Title')
plt.plot([x**2 for x in range(10)], label='$y=x^2$')
plt.legend(loc = 2)
```

## Several plots at one figure

#### Data generation

```python
linear = range(1, 11)
quadratic = [x**2 for x in range(1, 11)]
logarithmic = map(log, range(1, 11))
```

```python
from math import log
```

```python
print "linear:", linear
print "quadratic:", quadratic
print "logarithmic:", logarithmic
```

#### Making a plot

```python
plt.plot(linear, 'b-')
plt.plot(logarithmic, 'g-')
plt.plot(quadratic, 'r-')
```

#### Adding a legend

```python
plt.plot(linear, 'b-')
plt.plot(logarithmic, 'g-')
plt.plot(quadratic, 'r-')
plt.legend(["linear", "quadratic", "logarithmic"], loc = 2)
```

```python
plt.plot(linear, 'b-', label = "linear")
plt.plot(logarithmic, 'g-', label = "logarithmic")
plt.plot(quadratic, 'r-', label = "quadratic")
plt.legend(loc = 2)
```

#### Filling place between the lines

```python
plt.plot(linear, 'b-', linewidth=2.0, label = "linear")
plt.plot(logarithmic, 'g-', linewidth=2.0, label = "logarithmic")
plt.fill_between(range(10), logarithmic, linear, color='blue', alpha=0.2, label='place between')
plt.legend(loc = 2)
```

## Several plots at different figures

```python
plt.figure(figsize=(6,8))

plt.subplot(2,1,1)
plt.plot(linear, 'b-')
plt.title('Linear')

plt.subplot(2,1,2)
plt.plot(logarithmic, 'g-')
plt.title('Logarithmic')
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(linear, 'b-')
plt.title('Linear')

plt.subplot(1,2,2)
plt.plot(logarithmic, 'g-')
plt.title('Logarithmic')
```
