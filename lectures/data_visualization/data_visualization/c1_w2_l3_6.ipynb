## Simple plots with plotly

### Data generation

```python
x = np.linspace(0, 20, 100)
sin_data = np.sin(x)
```

### Plotting the interactive graph

```python
trace = go.Scatter(
    y = sin_data
)
data = [trace]
py.iplot(data)
```

### Plot with the datest

```python
x = pd.date_range(start = datetime.datetime.now(), end = datetime.datetime.now() + datetime.timedelta(100))
```

```python
trace = go.Scatter(
    x = x,
    y = sin_data
)
data = [trace]
py.iplot(data)
```

### Scatter plot

```python
trace = go.Scatter(
    x = x,
    y = sin_data,
    mode = 'markers'
)
data = [trace]
py.iplot(data)
```

### Plot properties

#### Color and mode

```python
trace = go.Scatter(
    x = x,
    y = sin_data,
    mode = 'lines+markers',
    line = dict(
        color=('red')
    )
)
data = [trace]
py.iplot(data)
```

#### RGB color and line width

```python
trace = go.Scatter(
    x = x,
    y = sin_data,
    mode = 'lines+markers',
    line = dict(
        color='rgb(30, 185, 18)',
        width=4.5
    )
)
data = [trace]
py.iplot(data)
```

### Several plots at one figure

```python
first_trace = go.Scatter(
    x = x,
    y = sin_data,
    mode = 'lines+markers',
    line = dict(
        color='yellow',
        width=3
    )
)

second_trace = go.Scatter(
    x = x,
    y = sin_data * 2,
    mode = 'markers',
    line = dict(
        color='rgb(70, 95, 218)',
        width=2
    )
)
data = [first_trace, second_trace]
py.iplot(data)
```

#### Label setting

```python
first_trace = go.Scatter(
    x = x,
    y = sin_data,
    mode = 'lines+markers',
    line = dict(
        color='black',
        width=3
    ),
    name = 'line'
)

second_trace = go.Scatter(
    x = x,
    y = sin_data * 2,
    mode = 'markers',
    line = dict(
        color='rgb(70, 95, 218)',
        width=2
    ),
    name = 'markers'
)
data = [first_trace, second_trace]


layout = dict(title = 'Plotly plot',
              xaxis = dict(title = 'DateTime'),
              yaxis = dict(title = 'Value'))

fig = dict(data=data, layout=layout)
py.iplot(fig)
```

### Bar plot

```python
data = [go.Bar(
            x=[20, 14, 23],
            y=['cats', 'dogs', 'monkeys'],
            orientation = 'h'
)]

py.iplot(data)
```

### Scatter plot

```python
trace = go.Scatter(
    y = np.random.randn(500),
    mode='markers',
    marker=dict(
        size='16',
        color = np.random.randn(500), #setting color equal to a variable
        colorscale='Viridis',
        showscale=True
    )
)
data = [trace]

py.iplot(data)
```

### Histogramm plot

```python
# Histogram data generation
x1 = np.random.randn(200) - 2  
x2 = np.random.randn(200)  
x3 = np.random.randn(200) + 2  
x4 = np.random.randn(200) + 4  

hist_data = [x1, x2, x3, x4]

group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Creation of the distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)

# Plotting
py.iplot(fig)
```

Еще одной интересной особенностью plotly является возможность быстро поделиться интерактивным графиком (в html формате).

Для этого достаточно вызвать метод plot, вместо iplot. Вы увидите путь до созданного файла, который можно будет просмотреть в любом браузере.

### Saving an interactive plot into html file

```python
# Histogram data generation
x1 = np.random.randn(200)-2  
x2 = np.random.randn(200)  
x3 = np.random.randn(200)+2  
x4 = np.random.randn(200)+4  

hist_data = [x1, x2, x3, x4]

group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Creation of the distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)

# Plotting
py.plot(fig, filename='dist_plot.html')
```
